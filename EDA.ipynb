# Exploratory Data Analysis

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cycles_df = pd.read_csv('Cleaned-Data/cleaned-physiological-cycles.csv')
cycles_df.head()
```

```python
cycles_df.describe()
```

```python
cycles_df.shape
```

```python
cycles_df.isnull().sum()
```

Making sure everything from the preprocessing is good. 

```python
# Plotting the distribution of Recovery Scores
plt.figure(figsize=(8, 6))
plt.hist(cycles_df['Recovery score %'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Recovery Scores')
plt.xlabel('Recovery Scores')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### The Effect Strain has on Sleep

```python
plt.figure(figsize=(12, 6))  

plt.subplot(1, 2, 1)
plt.scatter(cycles_df['Day Strain'], cycles_df['Asleep duration (min)'], c='skyblue', marker='o', label='Data Points', edgecolor='black')
plt.xlabel('Strain')
plt.ylabel('Sleep Time')
plt.title('Scatter Plot of Strain vs. Sleep Time')

plt.subplot(1, 2, 2)
plt.scatter(cycles_df['Day Strain'], cycles_df['Sleep efficiency %'], c='skyblue', marker='o', label='Data Points', edgecolor='black')
plt.xlabel('Strain')
plt.ylabel('Sleep Efficiency')
plt.title('Scatter Plot of Strain vs. Sleep Efficiency')
```

```python
plt.figure(figsize=(18, 6))  

plt.subplot(1, 3, 1)
plt.scatter(cycles_df['Day Strain'], cycles_df['Light sleep duration (min)'], c='skyblue', marker='o', label='Data Points', edgecolor='black')
plt.xlabel('Strain')
plt.ylabel('Light Sleep (Min)')
plt.title('Scatter Plot of Strain vs. Light Sleep')

plt.subplot(1, 3, 2)
plt.scatter(cycles_df['Day Strain'], cycles_df['Deep (SWS) duration (min)'], c='skyblue', marker='o', label='Data Points', edgecolor='black')
plt.xlabel('Strain')
plt.ylabel('Deep Sleep')
plt.title('Scatter Plot of Strain vs. Deep Sleep')

plt.subplot(1, 3, 3)
plt.scatter(cycles_df['Day Strain'], cycles_df['REM duration (min)'], c='skyblue', marker='o', label='Data Points', edgecolor='black')
plt.xlabel('Strain')
plt.ylabel('REM Sleep (Min)')
plt.title('Scatter Plot of Strain vs. REM Sleep')
```

```python
correlation_matrix = cycles_df[['Day Strain', 'Asleep duration (min)', 'Sleep efficiency %','Light sleep duration (min)', 'Deep (SWS) duration (min)', 'REM duration (min)']].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

### The effect calories burned has on sleep

```python
# calories burned vs sleep time, calories burned vs sleep efficiency 
plt.figure(figsize=(12, 6))  

plt.subplot(1, 2, 1)
plt.scatter(cycles_df['Energy burned (cal)'], cycles_df['Asleep duration (min)'], c='skyblue', marker='o', label='Data Points', edgecolor='black')
plt.xlabel('Calories Burned')
plt.ylabel('Sleep Time')
plt.title('Scatter Plot of Calories Burned vs. Sleep Time')

plt.subplot(1, 2, 2)
plt.scatter(cycles_df['Energy burned (cal)'], cycles_df['Sleep efficiency %'], c='skyblue', marker='o', label='Data Points', edgecolor='black')
plt.xlabel('Calories Burned')
plt.ylabel('Sleep Efficiency')
plt.title('Scatter Plot of Calories Burned vs. Sleep Efficiency')
```

```python
plt.figure(figsize=(18, 6))  

plt.subplot(1, 3, 1)
plt.scatter(cycles_df['Energy burned (cal)'], cycles_df['Light sleep duration (min)'], c='skyblue', marker='o', label='Data Points', edgecolor='black')
plt.xlabel('Calories Burned')
plt.ylabel('Light Sleep (Min)')
plt.title('Scatter Plot of Calories Burned vs. Light Sleep')

plt.subplot(1, 3, 2)
plt.scatter(cycles_df['Energy burned (cal)'], cycles_df['Deep (SWS) duration (min)'], c='skyblue', marker='o', label='Data Points', edgecolor='black')
plt.xlabel('Calories Burned')
plt.ylabel('Deep Sleep')
plt.title('Scatter Plot of Calories Burned vs. Deep Sleep)')

plt.subplot(1, 3, 3)
plt.scatter(cycles_df['Energy burned (cal)'], cycles_df['REM duration (min)'], c='skyblue', marker='o', label='Data Points', edgecolor='black')
plt.xlabel('Caloires Burned')
plt.ylabel('REM Sleep (Min)')
plt.title('Scatter Plot of Calories Burned vs. REM Sleep')
```

```python
correlation_matrix = cycles_df[['Energy burned (cal)', 'Asleep duration (min)', 'Sleep efficiency %','Light sleep duration (min)', 'Deep (SWS) duration (min)', 'REM duration (min)']].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

No correlation between Strain and Sleep and Calories Burned and Sleep

```python
plt.figure(figsize=(8, 6))
plt.scatter(cycles_df['Recovery category'], cycles_df['Energy burned (cal)'], color='skyblue', edgecolor='black')
plt.title('Recovery category vs. Calories Burned')
plt.xlabel('Recovery Scores')
plt.ylabel('Calories Burned')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
