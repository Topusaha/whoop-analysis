### Handling Null Values

```python
cycles_df.isnull().sum()
```

```python
cycles_df.drop(['Skin temp (celsius)', 'Blood oxygen %'], axis=1, inplace=True) # dropping skin temp and blood oxygen cause theres no data in those columns 
```

```python
cycles_df['Sleep consistency %']
```

```python
cycles_df.tail()
```

```python
"""
Seems to be that the last row was when I took off my whoop and stopped using it which skews the values. It's best to take that row out. 
The other 3 rows that have a Null value for sleep consistency we can just add the medium value 
"""

cycles_df.drop(len(cycles_df)-1, axis=0, inplace=True)

imp = SimpleImputer(missing_values=np.nan, strategy='median')
sleep_consistency_new = imp.fit_transform(cycles_df['Sleep consistency %'].to_numpy().reshape(-1,1))
cycles_df['Sleep consistency %'] = sleep_consistency_new
cycles_df.tail()
```

```python
cycles_df.isnull().sum()
```

```python
cycles_df.head(2) 
```

Looks like when I first got the whoop it did not have data for some columns. I think it would be best to delete this column as well.

```python
cycles_df.drop(0, axis=0, inplace=True)
cycles_df.head(2)
```

```python
cycles_df.isnull().sum()
```

Perfect, No more null values in the dataset
